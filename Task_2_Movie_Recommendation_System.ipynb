```python
pivot = df_merged.pivot_table(index = 'user_id',columns = 'title', values = 'rating')
pivot.head()
```

```python
def movie_correlation(movie_title, n):
    movie_rating = pivot[movie_title]
    similar_to_movie = pivot.corrwith(movie_rating)
    corr_ = pd.DataFrame(similar_to_movie, columns = ['correlation'])
    corr_.dropna(inplace=True)
    return corr_.sort_values(by = 'correlation',ascending=False).head(n)
movie_correlation('12 Angry Men (1957)', n = 10)
```

```python
def recommend_movies_for_new_movie(new_movie_title, pivot, correlation_matrix, n=5):
    # Find correlation for the new movie
    new_movie_correlation = correlation_matrix[new_movie_title].dropna()

    # Exclude the new movie itself
    new_movie_correlation = new_movie_correlation[new_movie_correlation.index != new_movie_title]

    # Sort movies by correlation in descending order
    correlated_movies = new_movie_correlation.sort_values(ascending=False)

    # Get the top N correlated movies
    top_recommendations = correlated_movies.head(n)

    return top_recommendations
```

```python
def recommend_movies_for_new_movie(new_movie_title, pivot, correlation_matrix, n=5):
    # Try to find the exact match
    try:
        new_movie_correlation = correlation_matrix[new_movie_title].dropna()
    except KeyError:
        # If not found, try to find a case-insensitive match
        matching_titles = [title for title in correlation_matrix.index if title.lower() == new_movie_title.lower()]
        if not matching_titles:
            return f"No correlation information found for '{new_movie_title}'."

        # Use the first matching title found
        new_movie_correlation = correlation_matrix.loc[matching_titles[0]].dropna()

    # Exclude the new movie itself
    new_movie_correlation = new_movie_correlation[new_movie_correlation.index != new_movie_title]

    # Sort movies by correlation in descending order
    correlated_movies = new_movie_correlation.sort_values(ascending=False)

    # Get the top N correlated movies
    top_recommendations = correlated_movies.head(n)

    return top_recommendations
```
